```python
import pandas as pd
```

```python
FIND = "¶contract¶"


with open('result.csv', 'w', newline='', encoding='utf-8') as file_in:  
    with open('work.txt') as file:
        for row in file:
            if FIND in row:
                #print(row.replace('¶',','))
                row_res = row.replace('¶',',')
                file_in.write(row_res)  
# Close the file object
file_in.close()         
            
            
```
